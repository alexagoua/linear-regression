# VEILLE SUR PYPLOT

## INTRODUCTION

L'augmentation du volume de données produit par l'homme, a conduit les développeurs à se surpasser afin de mieux les traiter. Parmi ces outils on dénote la création et la vulgarisation de langage de programmation en statistiques tels que R, PYTHON etc... .
Cependant, pour être plus fonctionnels et efficaces ces langages ont besoin d'être continuellement améliorer et étoffés en librairies performantes. 
Ainsi, afin de mieux visualiser les jeux de données, la librairie  Matplotlib a été créée.

### <em>MATPLOTLIB</em>

Matplotlib est une bibliothèque du langage de programmation Python destinée à tracer et visualiser des données sous formes de graphiques (<a href='https://fr.wikipedia.org/wiki/Matplotlib'>Wikipedia</a>). Il affiche les données en 2D et est multiplateforme.Matplotlib, dans sa conception est basé sur le framework de Numpy et Scipy.

Il a été conçu par John Hunter en 2002, à l'origine comme un correctif à IPython pour permettre un traçage interactif de style MatLab via gnuplot à partir de la ligne de commande IPython. Fernando Perez <em>(co-créateur à l'époque)</em> se démenait à l'époque pour terminer son doctorat et a fait savoir à John qu'il n'aurait pas le temps de revoir le patch pendant plusieurs mois. John a pris cela comme un signal pour se lancer seul, et le package matplotlib est né, avec la version 0.1 publiée en 2003 et en est à la version 3.4.2 actuellement. 

Il a reçu un coup de pouce lorsqu'il a été adopté comme package de traçage de choix du <strong>Space Telescope Science Institute</strong> de la NASA, qui a soutenu financièrement le développement de matplotlib et conduit à des capacités considérablement élargies (<a href='https://jakevdp.github.io/blog/2013/03/23/matplotlib-and-the-future-of-visualization-in-python/'>Matplotlib and the Future of Visualization in Python</a>, 2013). Bien qu'il ait pour origine l'émulation des commandes graphiques MATLAB, il est indépendant de MATLAB et peut être utilisé de manière Pythonic, orientée objet. Bien que Matplotlib soit écrit principalement en Python pur, il fait un usage intensif de NumPy et d'autres codes d'extension pour fournir de bonnes performances, même pour les grands tableaux(<a href='https://matplotlib.org/stable/users/history.html'>Site Matplotlib<a/> consulté 15  juillet 2021).

Matplotlib est conçu avec la philosophie selon laquelle vous devriez être capable de créer des tracés simples avec seulement quelques commandes, ou une seule ! Si vous voulez voir un histogramme de vos données, vous ne devriez pas avoir besoin d'instancier des objets, d'appeler des méthodes, de définir des propriétés, etc. ça devrait juste fonctionner. 

La facilitation de tracé avec matplotlib est possible par l'utilisation du module <strong><em>PYPLOT</em></strong>.

### <em>PYPLOT</em>

Matplotlib.pyplot est une collection de fonctions qui permet à matplotlib de fonctionner comme MATLAB. Chaque fonction pyplot effectue des changements sur une figure. 
Ex: créer une figure, créer une zone de traçage, traçer des lignes dans une zone de trçage, décorer une courbe avec des étiquettes, etc... (<a href='https://matplotlib.org/stable/tutorials/introductory/pyplot.html'>Site Matplotlib<a/> consulté 15  juillet 2021).

### Importation des librairies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Test simple de la librairie sur une liste

```python
# Création d'une liste
plt.plot([1,2,3,4])
# labelisation de l'axe des ordonnées
plt.ylabel('quelques chiffres')
# Affichage du tracé
plt.show()
```

Nous observons ici une des forces de matplotlib.pyplot. À savoir l'affichage automatique du tracé sans que nous ayons spécifié les axes. Si vous fournissez une seule liste ou un seul tableau à tracer, matplotlib suppose qu'il s'agit d'une séquence de valeurs y et génère automatiquement les valeurs x pour vous. Étant donné que les plages python commencent par 0, le vecteur x par défaut a la même longueur que y mais commence par 0. Par conséquent, les données x sont [0, 1, 2, 3]

```python
plt.plot([1,2,3,4],[1,4,9,16])
```

Ici nous avons spécifié les axes sous la forme (x,y).

```python
plt.plot([1,2,3,4],[1,4,9,16], '--k')
plt.axis([0,6,0,20])
plt.show()
```

Cette fois, nous avons pas spécifié l'échelle pour les abscisses et ordonnées et appliqué un style au tracé.
